### drug bins:

soft (alcohol, caffeine, cannabis, nicotine)
prescription (amphetamines, benzodiazapenes)
stimulants (amphetamines, caffeine, cocaine, crack, ecstasy, meth)
depressants (alcohol, benzos, heroin, ketamine)
psychedelics (LSD, mushrooms, ketamine?)
legal (alcohol, caffeine, nicotine, amyl, legal highs, VSA)
illegal (cocaine, crack, ecstasy, heroin, LSD, meth, mushrooms)
inhalants (amyl, VSA)
drop whole row: semer
drop col: chocolate

```python
df["Soft"] = None
def calculate_soft(df):
    for index, row in df.iterrows():
        alcohol = int(row["Alcohol"])
        caff = int(row["Caff"])
        cannabis = int(row["Cannabis"])
        nicotine = int(row["Nicotine"])
        df.loc[index, "Soft"] = max(alcohol, caff, cannabis, nicotine)

calculate_soft(df)
```

```python
df
```

```python
df["Soft"].value_counts()
```

```python
pd.set_option('display.max_columns', None)
```

```python
df["Prescription"] = None
def calculate_prescription(df):
    for index, row in df.iterrows():
        amphet = int(row["Amphet"])
        benzos = int(row["Benzos"])
        df.loc[index, "Prescription"] = max(amphet, benzos)

calculate_prescription(df)
```

```python
df
```

```python
df["Prescription"].value_counts()
```

```python
df["Stims"] = None
def calculate_stims(df):
    for index, row in df.iterrows():
        amphet = int(row["Amphet"])
        caff = int(row["Caff"])
        coke = int(row["Coke"])
        crack = int(row["Crack"])
        xtc = int(row["Ecstasy"])
        meth = int(row["Meth"])
        df.loc[index, "Stims"] = max(amphet, caff, coke, crack, xtc, meth)

calculate_stims(df)
```

```python
df["Stims"].value_counts()
```

```python
df["HardStims"] = None
def calculate_hstims(df):
    for index, row in df.iterrows():
        amphet = int(row["Amphet"])
        coke = int(row["Coke"])
        crack = int(row["Crack"])
        xtc = int(row["Ecstasy"])
        meth = int(row["Meth"])
        df.loc[index, "HardStims"] = max(amphet, coke, crack, xtc, meth)

calculate_hstims(df)
```

```python
df["Downers"] = None
def calculate_downers(df):
    for index, row in df.iterrows():
        alc = int(row["Alcohol"])
        benzos = int(row["Benzos"])
        h = int(row["Heroin"])
        k = int(row["Ketamine"])
        df.loc[index, "Downers"] = max(alc, benzos, h, k)

calculate_downers(df)
```

```python
df["Downers"].value_counts()
```

```python
df["HardDowners"] = None
def calculate_hdowners(df):
    for index, row in df.iterrows():
        benzos = int(row["Benzos"])
        h = int(row["Heroin"])
        k = int(row["Ketamine"])
        df.loc[index, "HardDowners"] = max(benzos, h, k)

calculate_hdowners(df)
```

```python
df["HardDowners"].value_counts()
```

```python
df["Psychs"] = None
def calculate_psychs(df):
    for index, row in df.iterrows():
        lsd = int(row["LSD"])
        shrooms = int(row["Mushrooms"])
        k = int(row["Ketamine"])
        df.loc[index, "Psychs"] = max(lsd, shrooms, k)

calculate_psychs(df)
```

```python
df["Psychs"].value_counts()
```

```python
#legal (alcohol, caffeine, nicotine, amyl, legal highs, VSA)
df["Legal"] = None
def calculate_legal(df):
    for index, row in df.iterrows():
        alc = int(row["Alcohol"])
        caff = int(row["Caff"])
        nic = int(row["Nicotine"])
        amyl = int(row["Amyl"])
        lh = int(row["Legalh"])
        vsa = int(row["VSA"])
        df.loc[index, "Legal"] = max(alc, caff, nic, amyl, lh, vsa)

calculate_legal(df)
```

```python
df["Legal"].value_counts()
```

```python
#illegal (cocaine, crack, ecstasy, heroin, LSD, meth, mushrooms)

df["Illegal"] = None
def calculate_illegal(df):
    for index, row in df.iterrows():
        c = int(row["Coke"])
        crack = int(row["Crack"])
        xtc = int(row["Ecstasy"])
        h = int(row["Heroin"])
        lsd = int(row["LSD"])
        meth = int(row["Meth"])
        shrooms = int(row["Mushrooms"])
        df.loc[index, "Illegal"] = max(c, crack, xtc, h, lsd, meth, shrooms)

calculate_illegal(df)
```

```python
df["Illegal"].value_counts()
```

```python
#inhalants (amyl, VSA)
df["Inhalants"] = None
def calculate_inhalants(df):
    for index, row in df.iterrows():
        amyl = int(row["Amyl"])
        vsa = int(row["VSA"])
        df.loc[index, "Inhalants"] = max(amyl, vsa)

calculate_inhalants(df)
```

```python
df["Inhalants"].value_counts()
```
