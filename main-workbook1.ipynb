```python
df["Soft"] = None
def calculate_soft(df):
    for index, row in df.iterrows():
        alcohol = int(row["Alcohol"])
        caff = int(row["Caff"])
        cannabis = int(row["Cannabis"])
        nicotine = int(row["Nicotine"])
        df.loc[index, "Soft"] = max(alcohol, caff, cannabis, nicotine)

calculate_soft(df)
```

```python
df["Prescription"] = None
def calculate_prescription(df):
    for index, row in df.iterrows():
        amphet = int(row["Amphet"])
        benzos = int(row["Benzos"])
        df.loc[index, "Prescription"] = max(amphet, benzos)

calculate_prescription(df)
```

```python
df["Stims"] = None
def calculate_stims(df):
    for index, row in df.iterrows():
        amphet = int(row["Amphet"])
        caff = int(row["Caff"])
        coke = int(row["Coke"])
        crack = int(row["Crack"])
        xtc = int(row["Ecstasy"])
        meth = int(row["Meth"])
        df.loc[index, "Stims"] = max(amphet, caff, coke, crack, xtc, meth)

calculate_stims(df)
```

```python
df["Psychs"] = None
def calculate_psychs(df):
    for index, row in df.iterrows():
        lsd = int(row["LSD"])
        shrooms = int(row["Mushrooms"])
        k = int(row["Ketamine"])
        df.loc[index, "Psychs"] = max(lsd, shrooms, k)

calculate_psychs(df)
```

```python
#legal (alcohol, caffeine, nicotine, amyl, legal highs, VSA)
df["Legal"] = None
def calculate_legal(df):
    for index, row in df.iterrows():
        alc = int(row["Alcohol"])
        caff = int(row["Caff"])
        nic = int(row["Nicotine"])
        amyl = int(row["Amyl"])
        lh = int(row["Legalh"])
        vsa = int(row["VSA"])
        df.loc[index, "Legal"] = max(alc, caff, nic, amyl, lh, vsa)

calculate_legal(df)
```

```python
df["Legal"].value_counts()
```

```python
#illegal (cocaine, crack, ecstasy, heroin, LSD, meth, mushrooms)

df["Illegal"] = None
def calculate_illegal(df):
    for index, row in df.iterrows():
        c = int(row["Coke"])
        crack = int(row["Crack"])
        xtc = int(row["Ecstasy"])
        h = int(row["Heroin"])
        lsd = int(row["LSD"])
        meth = int(row["Meth"])
        shrooms = int(row["Mushrooms"])
        df.loc[index, "Illegal"] = max(c, crack, xtc, h, lsd, meth, shrooms)

calculate_illegal(df)
```

```python
df["Illegal"].value_counts()
```

```python
#inhalants (amyl, VSA)
df["Inhalants"] = None
def calculate_inhalants(df):
    for index, row in df.iterrows():
        amyl = int(row["Amyl"])
        vsa = int(row["VSA"])
        df.loc[index, "Inhalants"] = max(amyl, vsa)

calculate_inhalants(df)
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import make_scorer, accuracy_score
from sklearn.multioutput import MultiOutputClassifier
from sklearn.preprocessing import MinMaxScaler

# Convert y_train and y_test to numpy arrays
y_train_array = y_train.values if isinstance(y_train, pd.DataFrame) else y_train
y_test_array = y_test.values if isinstance(y_test, pd.DataFrame) else y_test

# Define the model
base_rf = RandomForestClassifier(random_state=42)
model = MultiOutputClassifier(base_rf)

# Define the hyperparameter grid
param_grid = {
    'estimator__n_estimators': [100, 200, 500],
    'estimator__max_depth': [None, 10, 20],
    'estimator__min_samples_split': [2, 5, 10],
    'estimator__min_samples_leaf': [1, 2, 4]
}

# Define a custom scorer
def multioutput_accuracy(y_true, y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    accuracies = []
    for i in range(y_true.shape[1]):
        accuracies.append(accuracy_score(y_true[:, i], y_pred[:, i]))
    return np.mean(accuracies)

scorer = make_scorer(multioutput_accuracy, greater_is_better=True)

# Setup the grid search
grid_search = GridSearchCV(model, param_grid, scoring=scorer, cv=3, verbose=2, n_jobs=-1)

# Perform the grid search
grid_search.fit(X_train_norm, y_train_array)

# Print the best hyperparameters
print("Best hyperparameters found: ", grid_search.best_params_)

# Evaluate the best model on the test set
best_model = grid_search.best_estimator_
y_pred_test = best_model.predict(X_test_norm)

# Calculate and print the average accuracy for the test set
test_accuracies = []
for i in range(y_test_array.shape[1]):
    test_accuracies.append(accuracy_score(y_test_array[:, i], y_pred_test[:, i]))

average_test_accuracy = np.mean(test_accuracies)
print("Average Test Accuracy:", average_test_accuracy)
```
